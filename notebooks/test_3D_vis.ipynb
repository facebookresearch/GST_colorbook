```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Enable interactive mode
%matplotlib ipympl
#%matplotlib notebook

# Create some sample 3D data (replace with your own data)
n = 100  # Number of points
x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the scatter points
ax.scatter(x, y, z, c='b', marker='o')

# Set labels for the axes
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# The plot will now be interactive, allowing you to rotate, zoom, and pan.
```

```python
import torch

# Create a sample PyTorch tensor (replace with your own data)
n = 10  # Number of rows
m = 5   # Number of columns
tensor = torch.randn(n, m)

# Number of nearest elements to keep
k = 3

# Create a mask to keep the k nearest elements for each row
sorted_values, _ = torch.sort(tensor, dim=1)  # Sort each row along dimension 1
mask = (tensor <= sorted_values[:, k - 1:k])  # Select the k smallest values for each row

# Now, 'mask' is a boolean mask with True for the 30 nearest elements in each row.
```
